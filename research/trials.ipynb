```python
import random
import csv

def generate_random_rgb_and_names(num_lists=25):
    rgb_and_names = []

    for _ in range(num_lists):
        rgb_values = [random.randint(0, 255) for _ in range(3)]
        color_name = random.choice(['red', 'green', 'blue', 'yellow', 'purple'])  # Add more color names as needed
        rgb_and_names.append([rgb_values, color_name])

    return rgb_and_names

def save_to_csv(rgb_and_names, filename='colours.csv'):
    with open(filename, 'w', newline='') as csvfile:
        csv_writer = csv.writer(csvfile)
        csv_writer.writerow(['Red', 'Green', 'Blue', 'ColorName'])  # Header row

        for rgb_values, color_name in rgb_and_names:
            csv_writer.writerow(rgb_values + [color_name])

# Generate 25 lists of RGB colors with corresponding color names
rgb_and_names_list = generate_random_rgb_and_names()

# Save to CSV
save_to_csv(rgb_and_names_list)

# # Display the generated RGB and color names
# for rgb_values, color_name in rgb_and_names_list:
#     print(rgb_values, color_name)
```
